```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
Iris = list(uploaded.keys())[0]  # Gets the uploaded file name
df = pd.read_csv(Iris)
```

```python
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(12, 6))
features = ['Glucose', 'BMI', 'Age']
colors = ['blue', 'green', 'red']

for i, feature in enumerate(features):
    plt.hist(df[feature].dropna(), bins=20, alpha=0.5, color=colors[i], label=feature)

plt.title("Distribution of Glucose, BMI, and Age", fontsize=16)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['Glucose', 'BloodPressure', 'BMI', 'Age']], palette="Set2")
plt.title("Box Plot of Numerical Features", fontsize=16)
plt.show()
```

```python
sns.pairplot(df[['Glucose', 'BloodPressure', 'BMI', 'Age', 'Outcome']], hue='Outcome', palette='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x=df['Outcome'], palette="viridis")
plt.title("Diabetes Outcome Distribution", fontsize=16)
plt.xlabel("Outcome (0 = No Diabetes, 1 = Diabetes)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```
